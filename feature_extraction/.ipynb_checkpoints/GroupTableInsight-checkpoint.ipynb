```python
import pandas as pd
import numpy as np
```

```python
import data
```

```python
group_table, airline_table, order_table, train_set = data.import_data()
```

```python
main_table = pd.merge(train_set, order_table, on="order_id", how="left")
main_table = pd.merge(main_table, group_table, on="group_id", how="left")
```

```python
main_table["price_log"] = np.log(main_table["price"])
# 分別對成交與否的旅遊團價格畫histogram，發現差不多，因此未經處理的價格可能不是一個好的指標
pd.DataFrame.hist(main_table, column="price_log", by="deal_or_not")
```

```python
# 對每日的平均價格結果也差不多
main_table["day_price"] = main_table["price"] / main_table["days"]
main_table["day_price_log"] = np.log(main_table["day_price"])
main_table.hist(column="day_price_log", by="deal_or_not")
```

```python
# 旅行團的開始時間在 6, 7, 8, 9, 11, 12月的時候略高於其他月份
main_table["begin_date"] = pd.to_datetime(main_table["begin_date"])
main_table["begin_month"] = main_table["begin_date"].dt.month
pd.crosstab(main_table["begin_month"], main_table["deal_or_not"], normalize="index")
```

```python
# 一年當中的第200天左右出團的成交量比其他天高不少
main_table["begin_DoY"] = main_table["begin_date"].dt.dayofyear
main_table.hist(column="begin_DoY", by="deal_or_not")
```

```python
# week of year與day of year的結構上大抵相同
main_table["begin_WoY"] = main_table["begin_date"].dt.weekofyear
main_table.hist(column="begin_WoY", by="deal_or_not")
```

```python
# 旅遊團開始時間單看星期一到五似乎沒什麼區別，但是有可能跟其他變量會有交互作用
main_table["begin_DoW"] = main_table["begin_date"].dt.dayofweek
pd.crosstab(main_table["begin_DoW"], main_table["deal_or_not"], normalize="index")
```

```python
# 各area的成交與否結構明顯不同，但之前處理的時候直接把area_value最後一位當area，取錯
pd.crosstab(main_table["area"], main_table["deal_or_not"], normalize="index")
```

```python
# 單看subline，結構大致上相同，然而subline_value_3的成交率居然高達0.6
# 不過subline與area應該要組合在一起看
pd.crosstab(main_table["sub_line"], main_table["deal_or_not"], normalize="index")
```
