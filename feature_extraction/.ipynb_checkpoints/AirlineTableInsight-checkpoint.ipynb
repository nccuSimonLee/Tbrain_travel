```python
import pandas as pd
import numpy as np
from pyairports.airports import Airports
```

```python
airports = Airports()
```

```python
airports.airport_iata("FCO")
```

```python
import data
```

```python
group_table, airline_table, order_table, train_set = data.import_data()
```

```python
# 計算飛行時長，但部分資料計算出來的飛行時長是負數
airline_table["fly_time"] = pd.to_datetime(airline_table["fly_time"])
airline_table["arrive_time"] = pd.to_datetime(airline_table["arrive_time"])
for tmp in ["src_airport", "dst_airport"]:
    airport_split = airline_table[tmp].str.split(" ", expand=True)
    airline_table[tmp + "_code"] = airport_split[0]
    airline_table[tmp + "_name"] = airport_split[1]
airline_table.drop(columns=["src_airport", "dst_airport"], inplace=True)
x = np.array(list(map(lambda x: float(airports.airport_iata(x)[-3]), airline_table["src_airport_code"])))
y = np.array(list(map(lambda x: float(airports.airport_iata(x)[-3]), airline_table["dst_airport_code"])))
airline_table["jet_lag"] = list((y - x))

airline_table["fly_duration"] = list(map(lambda x: x.total_seconds(), airline_table["arrive_time"] - airline_table["fly_time"]))
airline_table["fly_duration"] = (airline_table["fly_duration"] - (airline_table["jet_lag"] * 60 * 60)) / (60 * 60)
x = airline_table[airline_table["fly_duration"] < 0 
```

```python
airline_table["fly_time"] = pd.to_datetime(airline_table["fly_time"])
airline_table["arrive_time"] = pd.to_datetime(airline_table["arrive_time"])
```

```python
# 旅行團會經過的機場數量
def airport_amount(tmp_table):
    return len(set(tmp_table["src_airport"]).union(tmp_table["dst_airport"]))
```

```python
# 出國時的機場與回國時的機場是否一致
def airport_consistent(tmp_table):
    if tmp_table["src_airport"].iloc[0] == tmp_table["dst_airport"].iloc[-1]:
        return 1
    else:
        return 0
```

```python
new_airport_table = pd.DataFrame(columns=["group_id"])
new_airport_table["group_id"] = list(set(airline_table["group_id"]))
x = airline_table.groupby(["group_id"])
new_airport_table["airport_amount"] = x.apply(airport_amount).tolist()
new_airport_table["airport_consistent"] = x.apply(airport_consistent).tolist()
new_airport_table["abroad_airport"] = x.apply(lambda x: x["src_airport"].iloc[0]).tolist()
new_airport_table["abroad_time"] = x.apply(lambda x: x["fly_time"].iloc[0]).tolist()
new_airport_table["home_airport"] = x.apply(lambda x: x["dst_airport"].iloc[-1]).tolist()
new_airport_table["home_time"] = x.apply(lambda x: x["arrive_time"].iloc[-1]).tolist()
```

```python
def test_func(tmp_table):
    if tmp_table["src_airport"].iloc[0] == 'HKG 香港機場':
        return tmp_table["group_id"].iloc[0]
    else:
        return 0

abnormal_list = x.apply(test_func).tolist()
```

```python
new_airport_table["abroad_hour"] = new_airport_table["abroad_time"].dt.hour
new_airport_table["home_year"] = new_airport_table["home_time"].dt.year
new_airport_table["home_DoW"] = new_airport_table["home_time"].dt.dayofweek
new_airport_table["home_DoY"] = new_airport_table["home_time"].dt.dayofyear
new_airport_table["home_hour"] = new_airport_table["home_time"].dt.hour
```

```python
main_table = pd.merge(train_set, order_table, on="order_id")
main_table = pd.merge(main_table, new_airport_table, on="group_id")
```

```python
# 途經機場數量與成交與否的結構差不多
pd.crosstab(main_table["airport_amount"], main_table["deal_or_not"], normalize="index")
```

```python
# 出國與回國時的機場一致與否 與 成交與否 的結構差不多
pd.crosstab(main_table["airport_consistent"], main_table["deal_or_not"], normalize="index")
```

```python
# 登機時間在午夜12點以後到凌晨5點之間的成交率偏低
pd.crosstab(main_table["abroad_hour"], main_table["deal_or_not"], normalize="index")
```

```python
# 回國落地時間與成交與否似乎沒有太大的關係
pd.crosstab(main_table["home_hour"], main_table["deal_or_not"], normalize="index")
```

```python
# 星期幾回到台灣與成交與否似乎沒有太大的關係
pd.crosstab(main_table["home_DoW"], main_table["deal_or_not"], normalize="index")
```

```python
# 7 8月左右回台灣成交率偏高，因為出國與回國之間相差不超過一個月，這個feature可能是redundancy
main_table.hist(column="home_DoY", by="deal_or_not")
```
